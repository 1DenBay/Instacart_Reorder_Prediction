```python
import sqlite3
import pandas as pd
import os

# Veritabanı yolunu dinamik olarak buluyoruz
DB_PATH = os.path.join("..", "data", "processed", "instacart.db")

def get_query(query):
    """
    Bu fonksiyona utility function (araç fonksiyonu) denir. yardımcı görevi görür
    SQL sorgusunu çalıştırır ve sonucu Pandas DataFrame olarak döndürür. 
    """
    # Bağlantıyı aç
    conn = sqlite3.connect(DB_PATH)
    
    # Sorguyu çalıştır ve veriyi çek
    df = pd.read_sql(query, conn)
    
    # Bağlantıyı kapat (Çok önemli! Yoksa veritabanı kilitli kalır)
    conn.close()
    
    return df

print("✅ Veritabanı bağlantı fonksiyonu hazır!")
```

```python
# Veritabanındaki tüm tabloların ismini getir
query = "SELECT name FROM sqlite_master WHERE type='table';"
get_query(query)
```

```python
# Orders tablosunun ilk 10 satırını getir 
query = """
SELECT * FROM orders 
LIMIT 10;
"""
get_query(query)
```

```python
# toplam kaç kullanıcı ve kaç sipariş olduğunu getir
query = """ 
SELECT 
    COUNT(DISTINCT user_id) as total_users,
    COUNT(order_id) as total_orders
FROM orders;
"""
get_query(query)
```

```python
# ürünlerin isimlerini products tablosundn, sipariş detaylarını order_products__prior tablosundan alarak en çok satın alınan 5 ürünü getir
query = """
SELECT 
    p.product_name,
    COUNT(*) as buying_count
FROM order_products__prior op
JOIN products p ON op.product_id = p.product_id
GROUP BY p.product_name
ORDER BY buying_count DESC
LIMIT 5;
"""
get_query(query)
```
